```python
text = "La UP es el primer lugar entre las universidades privadas"
matchObj = re.match(r"(.*) entre (.*?) .*", text, re.M|re.I)

if matchObj:
    print("matchObj.group() : ", matchObj.group())
    print("matchObj.group(1) : ", matchObj.group(1))
    print("matchObj.group(2) : ", matchObj.group(2))
else:
    print("No match!!")
```

```python
matchObj = re.match(r"UP", text, re.M|re.I)
if matchObj:
    print("match --> matchObj.group(): ", matchObj.group())
else:
    print("No match!!")

matchObj = re.search(r"UP", text, re.M|re.I)
if matchObj:
    print("match --> matchObj.group(): ", matchObj.group())
else:
    print("No match!!")
```
